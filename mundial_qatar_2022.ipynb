## Analisis

### Primera Parte

```python
# Cuantas selecciones participaron?
# Completar
# Cuantas selecciones participaron?
todos_los_equipos = set(mundial['1'].unique()) | set(mundial['2'].unique())
Total_Selecciones = len(todos_los_equipos)
print("Selecciones participantes:", Total_Selecciones)
```

```python
# Cuantas fases tiene el mundial?
# Completar
fases = mundial["group"].unique()
print("Fases:", len(fases))
print(fases)
```

```python
# Cual fue la "attendance" promedio por partido?
# Completar
print("Attendance promedio:", mundial["attendance"].mean())
```

```python
# Cual fue el partido con mas diferencia de goles?
# Completar
mundial["diferencia_goles"] = abs(mundial["1_goals"] - mundial["2_goals"])
idx = mundial["diferencia_goles"].idxmax()
p = mundial.loc[idx]
print(f"{p['1']} {p['1_goals']} - {p['2_goals']} {p['2']}  (diferencia: {p['diferencia_goles']})")
```

```python
# Cual es la selección que mas penales metió?
# Completar
pen_local     = mundial.groupby("1")["1_panelties_scored"].sum()
pen_visitante = mundial.groupby("2")["2_panelties_scored"].sum()
pen_total = pen_local.add(pen_visitante, fill_value=0)
print("Más penales metió:", pen_total.idxmax(), "-", int(pen_total.max()))
```

```python
# Cual es la selección que mas penales le hicieron?
# Completar
rec_local     = mundial.groupby("1")["2_panelties_scored"].sum()  # local recibe los del visitante
rec_visitante = mundial.groupby("2")["1_panelties_scored"].sum()  # visitante recibe los del local
rec_total = rec_local.add(rec_visitante, fill_value=0)
print("Más penales le hicieron a:", rec_total.idxmax(), "-", int(rec_total.max()))
```

### Segunda Parte

```python
equipo : str = "ARGENTINA"
```

```python
# Seleccionar todos los partidos que jugo el equipo de "local"
# Completar

filtroLocal : pd.Series = mundial["1"] == equipo
```

```python
# Seleccionar todos los partidos que jugo el equipo de "visitante"
# Completar

filtroVisitante : pd.Series = mundial["2"] == equipo
```

```python
# Seleccionar todos los partidos que jugo el equipo (de "local" o "visitante")
# Completar

partidos : pd.DataFrame = mundial[filtroLocal | filtroVisitante]
```

```python
# Cuantos goles hizo el equipo en total (obtener los goles metidos de local y sumarlos con los goles metidos de visitante)
# Completar

goles_local = partidos[partidos["1"] == equipo]["1_goals"].sum()
goles_visitante = partidos[partidos["2"] == equipo]["2_goals"].sum()
goles : int = goles_local + goles_visitante
print("Goles totales:", goles)
```

```python
# Cuantos goles hizo el equipo en promedio
# Completar
print("Goles promedio:", goles / len(partidos))
```

```python
# Cuantos partidos gano el equipo
# Completar

ganados_local = partidos[(partidos["1"] == equipo) & (partidos["1_goals"] > partidos["2_goals"])]
ganados_visitante = partidos[(partidos["2"] == equipo) & (partidos["2_goals"] > partidos["1_goals"])]
ganados : int = len(ganados_local) + len(ganados_visitante)
print("Partidos ganados:", ganados)
```

```python
# Graficar los goles hechos por el equipo en cada partido

fig, ax = plt.subplots()

goles_por_partido = []
for i, row in partidos.iterrows():
    if row["1"] == equipo:
        goles_por_partido.append(row["1_goals"])
    else:
        goles_por_partido.append(row["2_goals"])

ax.bar(range(len(partidos)), goles_por_partido)
ax.set_xlabel("Partido")
ax.set_ylabel("Goles")
ax.set_title(f"Goles de {equipo} por partido")
plt.show()
# Completar
```

### Tercera Parte

```python
# Contar cuantos goles hizo cada equipo en total en el mundial
# Completar

goles_local = mundial.groupby("1")["1_goals"].sum()
goles_visitante = mundial.groupby("2")["2_goals"].sum()
goles_totales = goles_local.add(goles_visitante, fill_value=0).sort_values(ascending=False)

equipos = goles_totales.index.tolist()
goles = goles_totales.values.tolist()
print(goles_totales)
```

```python
# Graficar los goles por equipo en todo el mundial

fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(equipos, goles)
ax.set_xlabel("Equipo")
ax.set_ylabel("Goles")
ax.set_title("Goles por equipo - Qatar 2022")
plt.xticks(rotation=45, ha="right")
plt.show()
# Completar
```

```python
# Cual es el equipo que hizo mas goles en el mundial?
# Completar
print("Equipo con mas goles:", goles_totales.idxmax())
```

```python
# Por cada partido, cual fue la diferencia de goles?
# Completar
mundial["diferencia_goles"] = abs(mundial["1_goals"] - mundial["2_goals"])
print(mundial[["1", "2", "1_goals", "2_goals", "diferencia_goles"]])
```

```python
# Por cada partido, cual fue el equipo que gano?
# Completar
for i, row in mundial.iterrows():
    if row["1_goals"] > row["2_goals"]:
        ganador = row["1"]
    elif row["2_goals"] > row["1_goals"]:
        ganador = row["2"]
    else:
        ganador = "Empate"
    print(f"{row['1']} {row['1_goals']} - {row['2_goals']} {row['2']} = {ganador}")
```

```python
# En cada fase del mundial, cual fue el equipo que hizo mas goles?
# Completar

fases_limpias = ["Group Stage", "Round of 16", "Quarter-final", "Semi-Final", "Play-off for third place", "Final"]

for fase in fases_limpias:
    if fase == "Group Stage":
        partidos_fase = mundial[mundial["group"].str.contains("Group")]
    else:
        partidos_fase = mundial[mundial["group"] == fase]

    goles_local = partidos_fase.groupby("1")["1_goals"].sum()
    goles_visitante = partidos_fase.groupby("2")["2_goals"].sum()
    goles_fase = goles_local.add(goles_visitante, fill_value=0)

    if len(goles_fase) == 0:
        print(f"{fase} = sin datos")
    else:
        print(f"{fase} = {goles_fase.idxmax()} ({int(goles_fase.max())} goles)")
```

### Tercera Parte

```python
# Armar un gráfico a partir de algun nuevo análisis que se te ocurra
# Completar
promedios = mundial.groupby("group").apply(lambda x: (x["1_goals"] + x["2_goals"]).mean())

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(promedios.index, promedios.values)
ax.set_xlabel("Fase")
ax.set_ylabel("Promedio de goles por partido")
ax.set_title("Promedio de goles por partido en cada fase - Qatar 2022")
plt.xticks(rotation=45, ha="right")
plt.show()
```
